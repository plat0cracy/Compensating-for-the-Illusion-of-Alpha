```r
##Load libraries
## Load Libraries
library(readxl)      # import Excel files
library(dplyr)       # data manipulation
library(tidyr)       # data cleaning
library(ggplot2)     # plotting (if needed)
library(broom)       # clean regression output
library(lubridate)   # work with dates
library(knitr)       # tables in Rmd / HTML
library(readr)   # for read_csv()


```

```r
## Load Data
## Load Data (CSV)
port <- read_csv("6_Portfolios_2x3_Clean.csv")
ff   <- read_csv("F-F_Research_Data_Factors_Clean.csv")

```

```r
port <- port %>%
  mutate(
    Year = as.numeric(Year),
    Month = trimws(Month)
  )

ff <- ff %>%
  mutate(
    Year = as.numeric(Year),
    Month = trimws(Month)
  )

glimpse(port)
glimpse(ff)
```

```r
data_raw <- inner_join(port, ff, by = c("Year", "Month"))

glimpse(data_raw)
nrow(data_raw)
```

```r
data <- data_raw %>%
  filter(Year > 1989) 
  # filter(!(Year == 1990 & Month != "January"))

data <- data %>%
  rename(
    Mkt_RF = `Mkt-RF`
  )

glimpse(data)
```

```r
# Identify portfolio & factor columns
port_cols <- c("M1B1","M1B2","M1B3","M2B1","M2B2","M2B3")
factor_cols <- c("Mkt_RF","SMB","HML","RF")

# Convert to decimals
data[port_cols] <- lapply(data[port_cols], function(x) x / 100)
data[factor_cols] <- lapply(data[factor_cols], function(x) x / 100)
```

```r
for (p in port_cols) {
  data[[paste0(p, "_excess")]] <- data[[p]] - data$RF
}
```

```r
portfolios <- paste0(port_cols, "_excess")

capm_results <- lapply(portfolios, function(p) {
  lm(as.formula(paste0(p, " ~ `Mkt_RF`")), data = data)
})
names(capm_results) <- portfolios
```

```r
ff_results <- lapply(portfolios, function(p) {
  lm(as.formula(paste0(p, " ~ `Mkt_RF` + SMB + HML")), data = data)
})
names(ff_results) <- portfolios
```

```r
# CAPM outputs
capm_alpha <- sapply(capm_results, function(m) coef(m)[1])
capm_r2    <- sapply(capm_results, function(m) summary(m)$adj.r.squared)

# FF outputs
ff_alpha <- sapply(ff_results, function(m) coef(m)[1])
ff_r2    <- sapply(ff_results, function(m) summary(m)$adj.r.squared)
```

```r
comparison <- data.frame(
  Portfolio = portfolios,
  CAPM_Alpha = capm_alpha,
  FF_Alpha = ff_alpha,
  CAPM_R2 = capm_r2,
  FF_R2 = ff_r2
)

comparison
```
